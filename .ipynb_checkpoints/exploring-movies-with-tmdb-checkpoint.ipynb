### General Properties
Below we are going to read in the data file we are using as well as display some summary statistics before we get into cleaning it. This section will provide a bit more insight on the data we plan to analyze.

```python
# Load Data and display first few records
df = pd.read_csv('tmdb-movies.csv')
df.head(5)
```

With the initial view of the dataset, we can see that there are definitely columns that we will not use as they likely will not have much use for our analysis. We will be removing these columns shortly.



### Cleaning TMDb for Analysis

Now we will be cleaning the dataset to fit our needs

#### We will conduct the following steps to achieve this goal:
* Drop unneeded columns
* Remove duplicate columns
* Make release date a datetime field
* Make budget_adj, revenue_adj floats and use million as unit of measurement
* Address empty cells 



#### Removing Unneeded Columns
The first step we will take to clean our data is to drop unneeded columns. In our analysis, we will not need the following:
* id
* imdb_id
* cast
* homepage
* tagline
* keywords
* overview
* budget
* revenue

The reason why the budget and revenue columns are being dropped is due to the presence of the adjusted values that are included with the dataset. The *budget_adj* and *revenue_adj* columns reflect the orginal columns but account for inflation overtime. They are displayed in terms of 2010 dollars. link: https://www.kaggle.com/tmdb/tmdb-movie-metadata/discussion/45990


```python
df.columns
```

```python
# Columns we do not need
# Drop Columns
to_drop = ['id','imdb_id','cast','homepage','overview','tagline','keywords', 'budget','revenue']

df.drop(to_drop,inplace=True,axis=1)
```

```python
# Display first five records dataset after dropping columns
df.head(5)
```

```python
df.info()
```

#### Remove Duplicate Rows
The code cell below will first check the dataset for any duplicates then drop them if any are present.

```python
# Check duplicates
sum(df.duplicated())

#Drop any Duplicates
df= df.drop_duplicates()

# Recheck duplicates
sum(df.duplicated())
```

#### Fix Release Date column
The cell below will modify the data type of the *release_date* field from string to datetime

```python
# Convert release_date to datetime
df['release_date'] = pd.to_datetime(df['release_date'])
```

#### Shortening budget and revenue fields
The cells below shorten the values in the adjusted budget and revenue columns. New columns are added to replace these columns to depict that that value is represented in millions.

```python
df['budget_adj_in_millions'] = (df['budget_adj']/1000000).round(2)
df['revenue_adj_in_millions'] = (df['revenue_adj']/1000000).round(2)
```

```python
#Drop original budget_adj and revenue_adj columns
df.drop(['budget_adj','revenue_adj'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.info()
```

#### Addressing Empty, or Zeroed, Cells
Looking back over the summary statistics of the current dataframe (see code output from cell below), there are many cells that have "0" in place of a numerical value or NaN. These are found in the *runtime*, *budget_adj_in_millions*, and *revenue_adj_in_millions* fields. To address whether or not to impute values or drop the column(s) entirely, lets identify how much of the data is actually "missing."

```python
df.describe()
```

The summary statistics above indicate that over half of the adjusted budget and revenue data has been omitted from this dataset. Since this is the case, we cannot use these columns for the purpose of our analysis.

```python
# Drop budget_adj_in_millions and revenue_adj_in_millions
df.drop(['budget_adj_in_millions','revenue_adj_in_millions'],axis=1,inplace=True)
df.head(2)
```

Now let's take a look at the runtime column. As it stands, the runtime column has enough data to impute values where its respective cells contain zero. With the presence of existential outliers in the dataset, we will impute these zero values with the median runtime.

```python
# Check how many values have 0 listed as the runtime
df.query('runtime == 0')['runtime'].count()
```

```python
# Replace all zero values with median runtime
df['runtime'].replace(0,df['runtime'].median(),inplace=True)
```

```python
# Re-check how many values have 0 listed as the runtime
df.query('runtime == 0')['runtime'].count()
```

```python
df.describe()
```

```python
df.info()
```

We have cleaned our dataset so that it is ready for our analysis! After it is all said and done, we have 10 actionable columns to work with.

### Which Established Director Makes the Best Movies?
For this question, we will look at the 25 directors who've generated the most movies. We will be evaluating a director's success based on the average rating their movies get. Voting averages for movies are based on a scale of 1-10 and are dictated by the number of votes input by its viewerbase.

First, let's identify the 25 directors who have made the most movies. To start, we will get a list of these directors and then plot the information on a horizontal bar chart.

```python
# Count top 25 directors
top_25_dirs = df['director'].value_counts().nlargest(25)
top_25_dirs
```

```python
# Bar Chart for top 25 directors by number of movies produced
top_25_dirs.sort_values(ascending=True).plot(kind='barh',figsize=(10,10))
plt.title('Top 25 Directors by Movie Count')
plt.xlabel('Number of Movies Directed')
plt.ylabel('Name');
```

A filter must be applied to the dataset now to include only the directors listed above. To perform this function, a merge will be used to ensure only the directors listed above will be inlcuded in the new dataset. 

Helpful link: https://stackoverflow.com/questions/45615306/get-top-rows-from-column-value-count-with-pandas

```python
# Merge top 25 director series with Dataframe to get movies directed by only by top 25 directors
top_dir_df = pd.DataFrame({'director':top_25_dirs.index}).merge(df, how='left')

# Check that new df director value counts matches top 25 director series value counts
top_dir_df['director'].value_counts()
```

Now that we have the entire dataset where we want it, we should be able to answer our question: which one of these directors makes the best movies? 

```python
top_dir_df.groupby('director')['vote_average'].mean().sort_values(ascending=False)
```

```python
# Plot directors based on average movie rating
top_dir_df.groupby('director')['vote_average'].mean().sort_values(ascending=True).plot(kind='barh',figsize=(10,10));
plt.xticks([5, 5.5, 6, 6.5, 7])
plt.xlim(left=5)
plt.title('Top 25 Directors by Average Movie Rating')
plt.xlabel('Average Rating for All Movies')
plt.ylabel('Name');
```

```python
top_dir_df.groupby('director')['vote_average'].max().sort_values(ascending=False)
```

The data above indicates the following information:
* Woody Allen is the most established director with 45 movies directed 
* Martin Scorsese produces the best movies with an overall average movie rating of approximately 7.0, with Steven Spielberg following closely behind with an all-time average movie rating of 6.8
* Francis Ford Coppola directed the best movie out of the top 25 established directors which holds an average rating of 8.3

### How Do Movie Genres Since 2000 Compare with Each Other?
For this research question, we will look at movies with respect to their genre. This notebook will examine the following qualities from movies by genre dating back to 2000:

* Number of movies released per year
* Average ratings
* Average runtimes

The first thing we will do is create a new dataframe with movies produced in the year 2000 and on.

```python
# Create new df for movies from 2000 to recent
df_since_2000 = df.query('release_year >= 2000')

# Verify and plot number of movies released per year from 2000 to most recent
df_since_2000['release_year'].value_counts().sort_index().plot(kind='bar',figsize=(15,5))
plt.title('Movies Released pre Year, 2000-2015',size=18)
plt.xlabel('Year',size=14)
plt.xticks(rotation=300)
plt.ylabel('Count');
```

More movies are being produced year after year. The graph above that plots the new dataframe shows a steady increase in movies being made over the years.

#### Dividing Movies Into Genres
Next, we will split the new dataframe by genre. Through initial investigation of the dataset, movies may be listed having multiple genres in the *genres* column. Movies with multiple genres have them separated with the "|" delimiter in the dataframe. We need to separate these genres by grouping movies based on individual categories.

For the sake of this research question, movies with multiple genres are included in multiple categories. 

Helpful link: https://stackoverflow.com/questions/61601350/pandas-groubpy-on-substring-or-regex

```python
# Get movie counts of individual genres
genre_series = df_since_2000.groupby(df['genres'].str.split('|').str[0]).size().sort_values(ascending=False)
genre_series
```

```python
# Bar chart of number of movies per genre
genre_series.plot(kind='bar',figsize=(25,5))
plt.title('Movies Produced by Genre, 2000-2015',size=20)
plt.ylabel('Count',size=14)
plt.xlabel('Genre',size=14)
plt.xticks(rotation=315);
```

```python
# Pie chart of movies produced by genre as a percentage
genre_series.plot.pie(autopct="%.1f%%",explode=[0.03]*len(genre_series),figsize=(14,14),pctdistance=0.75);
plt.title('Percentage of Movies Produced By Genre, 2000-2015',size=20)
plt.ylabel('');
```

We can see that movies falling under the drama category have been the most popular by a wide margin and movies considered as dramas and/or comedies account for almost half of all movies produced from 2000 to 2015.


#### Visualizing Movies Over Time
This section will analyze trends on a year-by-year basis. We will look at how movies trended positively or negatively overtime.

We will first plot what types of movies were produced over time via seaborn's heatmap capability.

Helpful links: 
* https://datavizpyr.com/heatmaps-with-seaborn-in-python/
* https://matplotlib.org/3.3.1/gallery/images_contours_and_fields/image_annotated_heatmap.html

```python
# Create Heatmap of movies count by year per genre
# Group By Genre and Release Year
tmp_df = df_since_2000.groupby([df['genres'].str.split('|').str[0],'release_year'],sort=False)['release_year'].agg(['count'])
tmp_df.head()

heatmap_df = pd.pivot_table(tmp_df, values ='count',index=['genres'],columns=['release_year'])

fig, ax = plt.subplots(figsize=(15,12.5))
sns.heatmap(heatmap_df, ax=ax)
plt.xlabel('Year', size=14)
plt.ylabel('Genre',size=14)
plt.title('Movies Released Per Year By Genre',size=18);
```

Action, Drama, Comedy, Thriller, Horror, and Documentary genres all stand out here. The heatmap shows positive trends for all of these categories, as more movies are typically being produced year-over-year. Also notable in the opposite light are Western and Foreign movies. These two categories have multiple years where not a single movies was produced.

Recall the earlier bar chart of movies produced by year. It is interesting to see that the best year for movie production appears to be 2014. This is also where dramas, comedies, and thriller movies see their peak in the heatmap above.

Next, we will look at how movie genres have been rated over time. Let's begin with a plot to show the average rating for all movies each year.

```python
# Get Average Movie rating per genre over entire timespan
genre_ratings_series = df_since_2000.groupby(df['genres'].str.split('|').str[0])['vote_average'].mean().sort_values(ascending=False)
genre_ratings_series.plot(kind='bar',figsize=(25,5));
plt.title('Average Movie Rating By Genre, 2000-2015',size=18)
plt.xlabel('Genre',size=14)
plt.xticks(rotation=315)
plt.ylabel('Rating',size=14)
plt.ylim(bottom=5);
```

The next plot is a heatmap of how individual genres fared over time.

```python
# Create Heatmap of movies count by year per genre
# Group By Genre and Release Year
tmp_df = df_since_2000.groupby([df['genres'].str.split('|').str[0],'release_year'],sort=False)['vote_average'].agg(['mean'])
tmp_df.head()

heatmap_df = pd.pivot_table(tmp_df, values ='mean',index=['genres'],columns=['release_year'])

fig, ax = plt.subplots(figsize=(15,12.5))
sns.heatmap(heatmap_df, annot=True, ax=ax)
plt.xlabel('Year', size=14)
plt.ylabel('Genre',size=14)
plt.title('Average Movie Rating Per Year By Genre',size=18);
```

Interesting to note that horror movies seem to continuously get voted as the worst category every year. Also, aside from a couple of bad years, science fiction movies are not as bad as they initially appear.

Lastly, we will average plot runtimes for each genre.

```python
# plot average runtime per genre
avg_runtime_series = df_since_2000.groupby(df_since_2000['genres'].str.split('|').str[0])['runtime'].mean().sort_values(ascending=False)
avg_runtime_series.plot(kind='bar',figsize=(15,5));
plt.title('Average Runtime by Genre, 2000-2015', size=18)
plt.xlabel('Genre',size=14)
plt.xticks(rotation=315)
plt.ylim(bottom=5)
plt.ylabel('Runtime (in minutes)',size=14);
```

```python
# Create Heatmap of movies count by year per genre
# Group By Genre and Release Year
tmp_df = df_since_2000.groupby([df['genres'].str.split('|').str[0],'release_year'],sort=False)['runtime'].agg(['mean'])
tmp_df.head()

heatmap_df = pd.pivot_table(tmp_df, values ='mean',index=['genres'],columns=['release_year'])

fig, ax = plt.subplots(figsize=(20,12.5))
sns.heatmap(heatmap_df, annot=True, fmt='.1f',ax=ax)
plt.xlabel('Year', size=14)
plt.ylabel('Genre',size=14)
plt.title('Average Runtime Per Year By Genre',size=18);
```
